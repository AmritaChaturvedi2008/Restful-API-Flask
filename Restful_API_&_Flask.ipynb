#Restful API & Flask

ques 1. What is a RESTful API?


ans -
A RESTful API is an API that follows the REST (Representational State Transfer) principles.
It uses HTTP methods like GET, POST, PUT, DELETE to let clients interact with a server.




ques 2. What is Flask, and why is it popular for building APIs?

ans-
Flask is a lightweight Python web framework used to build web applications and APIs.

Why it’s popular for APIs:

It is simple and easy to learn.

It is lightweight and gives developers full control.

It supports quick development with minimal code.

It has many extensions for adding features when needed.

ques 3.:What are HTTP methods used in RESTful APIs?

ans-
The main HTTP methods used in RESTful APIs are:

GET – Retrieve data

POST – Create new data

PUT – Update existing data

DELETE – Remove data

PATCH – Partially update data

  ans -

ques 6.  How do you create a basic Flask application?

ans - we can  create a basic Flask app by importing Flask, creating an app object, and defining a route.

Example:

from flask import Flask
app = Flask(__name__)

@app.route("/")
def home():
    return "Hello, Flask!"

app.run()


ques 7. How do you return JSON responses in Flask?


ans -
Use jsonify() to return JSON data in Flask.

from flask import jsonify

return jsonify({"message": "Success"})


ques 8. How do you handle POST requests in Flask?


ans

Use request.method == "POST" and read data from request.form or request.json.

Example:

from flask import request

@app.route("/submit", methods=["POST"])
def submit():
    data = request.json
    return data


ques 9. How do you handle errors in Flask (e.g., 404)?

ans

Use error handlers with the @app.errorhandler() decorator.

Example:
 @app.errorhandler(404)
def not_found(e):
    return "Page not found", 404
